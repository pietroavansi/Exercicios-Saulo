```
# Defina sua função aqui
def eh_par(numero):
  return numero % 2 == 0

# Faça a lógica do input e chamada da função aqui

num = int(input())
if eh_par(num):
  print(f"O número {num} é par!")
else:
  print(f"O número {num} não é par!")
```
